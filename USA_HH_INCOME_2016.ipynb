```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv
sns.set(style='whitegrid')
```

```python
df=pd.read_csv(r'c:\Users\INNOVATION\Documents\household income.csv')
df.head()
```

```python
print('Columns:',df.columns)
```

```python
neighborhood_col='Neighborhood'
income_col='Estimate; Aggregate household income in the past 12 months (in 2015 Inflation-adjusted dollars)'
```

```python
df[income_col] = pd.to_numeric(df[income_col], errors='coerce')
df = df.dropna(subset=[income_col])
```

```python
# SCATTER PLOT 
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x=neighborhood_col, y=income_col)
plt.xticks(rotation=90)
plt.title('Scatterplot of Household Income by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Aggregate Household Income')
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM 
plt.figure(figsize=(10, 6))
sns.histplot(df[income_col], bins=20, kde=True)
plt.title('Histogram of Household Income')
plt.xlabel('Aggregate Household Income')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#BAR CHART: Top 10 Neighborhoods by Income
top_n = 10
top_df = df.sort_values(by=income_col, ascending=False).head(top_n)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_df, x=neighborhood_col, y=income_col, palette='viridis')
plt.xticks(rotation=45)
plt.title(f'Top {top_n} Neighborhoods by Household Income')
plt.xlabel('Neighborhood')
plt.ylabel('Aggregate Household Income')
plt.tight_layout()
plt.show()
```

```python
#BAR CHART: Bottom 10 Neighborhoods by Income
bottom_n = 10
bottom_df = df.sort_values(by=income_col, ascending=False).tail(bottom_n)

plt.figure(figsize=(12, 6))
sns.barplot(data=bottom_df, x=neighborhood_col, y=income_col, palette='viridis')
plt.xticks(rotation=45)
plt.title(f'Bottom {bottom_n} Neighborhoods by Household Income')
plt.xlabel('Neighborhood')
plt.ylabel('Aggregate Household Income')
plt.tight_layout()
plt.show()
```
